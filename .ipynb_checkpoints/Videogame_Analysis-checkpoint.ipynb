# 1.Which titles sold the most worldwide? <a id='1'></a>

```python
sales=df[['title','total_sales']]
```

```python
sales.sort_values(by='total_sales',ascending=False,inplace=True)
```

```python
sales['total_sales'].head(70)
```

```python
plt.figure(figsize=(15,15))
sns.barplot(data=sales.head(50),x='total_sales',y='title',orient='h',color='b')
plt.xlabel('Total Sales (in Million)')
plt.ylabel('Title of Game')
plt.title('Most Sold Title WorldWide')
plt.show()
```

# 2.Which year had the highest sales? Has the industry grown over time? <a id='2'></a>

```python
years=df[['Year','total_sales']]
```

```python
years[years['Year'] == 2008].head(50)
```

```python
y_sales=years.groupby(['Year']).agg('sum')
```

```python
y_sales.sort_values(by='total_sales',ascending=False,inplace=True)
```

```python
y_sales
```

```python
plt.figure(figsize=(7,5))
sns.lineplot(data=y_sales,x=y_sales.index,y='total_sales',label='Year wise Sales',color='orange')
plt.xlabel('Year')
plt.ylabel('Total Sales (in Million)')
plt.title('Games Sales Over The Years')
plt.grid()
plt.show()
```

```python
y_sales.plot(kind='barh',figsize=(9,15))
plt.xlabel('Total Sales (in Million)')
plt.xlabel('Yaer')
plt.title('Game Sales Over The Year')
plt.show()
```

# 3.Do any consoles seem to specialize in a particular genre? <a id='3'></a>

```python
console=df[['console','genre']]
```

```python
console.loc[console['console'] == '3DO']
```

```python
console_grp=console.groupby(['console','genre']).value_counts()
```

```python
console_grp=pd.DataFrame(console_grp)
```

```python
console_grp.reset_index(inplace=True)
```

```python
console_grp
```

```python
action_consoles=console_grp[console_grp['genre'] == 'Action']
```

```python
action_consoles.sort_values(by='count',ascending=False,inplace=True)
```

```python
action_consoles.reset_index(inplace=True)
```

```python
action_consoles
```

```python
adventure_consoles=console_grp[console_grp['genre'] == 'Adventure']
```

```python
adventure_consoles.reset_index(inplace=True)
```

```python
adventure_consoles.sort_values(by='count',ascending=False,inplace=True)
```

```python
adventure_consoles
```

```python
plt.figure(figsize=(13,13))

#plotting adventure console in bar plot
sns.barplot(data=adventure_consoles.head(15),x='console',y='count',color='b',label='Adventure Console',alpha=0.4)

#plotting action console in bar plot
sns.barplot(data=action_consoles.head(15),x='console',y='count',color='orange',label='Action Console',alpha=0.3)
plt.show()
```

# 4.What titles are popular in one region but flop in another?<a id='4'></a>

```python
region=df[['title','na_sales','jp_sales','pal_sales','other_sales','total_sales']]
```

```python
region.drop(region[region['total_sales'] <= 1].index,inplace=True)
```

```python
region
```

```python
plt.figure(figsize=(15,15))

sns.barplot(data=region.head(25),x='title',y='na_sales',label='North America Sales',alpha=0.9)

sns.barplot(data=region.head(25),x='title',y='jp_sales',label='Japan Sales',alpha=0.3)

plt.xticks(rotation=90)

plt.title('Game Sales in North America and Japan')

plt.xlabel('Game Title')
plt.ylabel('Game Sales (in Million)')

plt.show()
```

```python
plt.figure(figsize=(15,15))

sns.barplot(data=region.head(25),x='title',y='pal_sales',label='European & African Sales',alpha=0.5)

sns.barplot(data=region.head(25),x='title',y='other_sales',label='Rest of World Sales',alpha=0.7)

plt.xticks(rotation=90)

plt.title('Game Sales in Europe/Africa and Rest of the World')

plt.xlabel('Game Title')
plt.ylabel('Game Sales (in Million)')

plt.show()
```
